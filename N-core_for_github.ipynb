```python
import pandas as pd
from tqdm import tqdm
```

```python
def filter_reviews(data, chk):
    cnt = 1
    while True:
        # 소비자 및 아이템별 리뷰 개수 계산
        consumer_review_counts = data['user'].value_counts()
        item_review_counts = data['item'].value_counts()

        # N개 미만의 리뷰를 가진 소비자 및 아이템 추출
        consumers_to_remove = consumer_review_counts[consumer_review_counts < chk].index
        items_to_remove = item_review_counts[item_review_counts < chk].index
        print(f"Epoch: {cnt}")
        print(f"리뷰가 {chk}개 미만인 소비자 수: {len(consumers_to_remove)}")
        print(f"리뷰가 {chk}개 미만인 아이템 수: {len(items_to_remove)}")
        print(f"Length of Data: {len(data)}")
        # 더 이상 N개 미만의 리뷰를 가진 소비자나 아이템이 없으면 종료
        if len(consumers_to_remove) == 0 and len(items_to_remove) == 0:
            print
            break

        # 5개 미만의 리뷰를 가진 소비자 또는 아이템이 있는 데이터 제거
        data = data[~data['user'].isin(consumers_to_remove)]
        data = data[~data['item'].isin(items_to_remove)]

        cnt += 1

    return data

```
